<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

Una capa de convolución es un componente crucial en las redes neuronales convolucionales, usadas principalmente para procesar imágenes. Opera mediante los siguientes pasos:

1. **Filtros:** Utiliza matrices pequeñas de pesos para extraer características específicas de la entrada, como bordes o texturas.
2. **Convolución:** Desliza estos filtros sobre la entrada para producir mapas de características, destacando dónde se encuentran estas características.
3. **Función de Activación:** A menudo se aplica una función de activación (como ReLU) para introducir no linealidades y permitir el aprendizaje de patrones complejos.
4. **Agrupación (Opcional):** Puede seguirse de una operación de agrupación para reducir la dimensionalidad y resumir la presencia de características.

**Utilidad:** Las capas de convolución son fundamentales para detectar automáticamente características importantes en imágenes, hacer que la red sea eficiente y robusta a variaciones en la posición de los objetos, y permitir el aprendizaje de una jerarquía de características complejas. Esto las hace esenciales para tareas como el reconocimiento de imágenes y la detección de objetos.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

Una capa de max-pooling es un componente comúnmente utilizado en redes neuronales convolucionales (CNNs) después de las capas de convolución. Su función principal es reducir la dimensionalidad de los mapas de características, lo que hace al modelo más eficiente y reduce el riesgo de sobreajuste. El procedimiento y la utilidad de una capa de max-pooling se pueden resumir en los siguientes puntos:

### Procedimiento:
1. **División en regiones:** La capa de max-pooling divide el mapa de características de entrada en regiones rectangulares no superpuestas. El tamaño de estas regiones se define por el tamaño de la ventana de pooling, que es un hiperparámetro (por ejemplo, 2x2).

2. **Extracción del valor máximo:** Para cada una de estas regiones, la capa de max-pooling selecciona el valor máximo, ignorando el resto de los valores.

3. **Formación del nuevo mapa de características:** Los valores máximos extraídos de todas las regiones forman un nuevo mapa de características. Este mapa será de menor tamaño que el original debido a la reducción de dimensionalidad.

### Utilidad:
- **Reducción de Dimensionalidad:** La capa de max-pooling reduce el número de parámetros y la cantidad de cómputo en la red, lo que ayuda a hacer el entrenamiento más eficiente y a disminuir el riesgo de sobreajuste.
  
- **Invariancia a pequeñas traslaciones:** Al seleccionar solo el valor máximo dentro de una ventana, la capa de max-pooling proporciona a la red una cierta invariancia a las traslaciones y distorsiones menores en la entrada, lo cual es útil en tareas de reconocimiento de patrones donde la posición exacta de las características no es crucial.
  
- **Enfoque en características relevantes:** Al conservar solo los valores máximos, se enfatizan las características más prominentes de la entrada, lo que puede mejorar la capacidad de la red para aprender patrones importantes y mejorar el rendimiento en tareas de clasificación y detección.



<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Una capa de "flatten" en redes neuronales convierte una estructura multidimensional de datos (como los mapas de características de una CNN) en un vector unidimensional. Este proceso es simple pero crucial para conectar capas convolucionales con capas densas (totalmente conectadas) en una red neuronal.

### Procedimiento:
- **Transformación:** Toma los mapas de características multidimensionales (por ejemplo, en 2D para imágenes) y los "aplana" en un único vector largo. Si tienes un mapa de características de tamaño [altura x anchura x canales], la capa de flatten reorganiza estos datos en un vector de tamaño [altura * anchura * canales].

### Utilidad:
- **Conexión con Capas Densas:** Las redes neuronales convolucionales (CNNs) suelen requerir de capas densas hacia el final para realizar tareas como clasificación. La capa de flatten es el puente que permite pasar de la representación espacial de las imágenes a una representación que las capas densas pueden procesar.
- **Preparación para Clasificación:** Al convertir los datos a un formato vectorial, facilita la tarea de las capas densas de realizar clasificaciones o regresiones basándose en las características extraídas por las capas convolucionales y de pooling.



<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

Para calcular el número total de parámetros en una red neuronal convolucional (CNN) con la estructura dada, debemos considerar tanto los pesos de los filtros en cada capa convolucional como los sesgos asociados a cada mapa de características. La fórmula general para calcular el número de parámetros en una capa convolucional es:

\[ \text{Número de parámetros} = \left( \text{Tamaño del filtro}^2 \times \text{Número de canales de entrada} + 1 \right) \times \text{Número de filtros} \]

Donde:
- **Tamaño del filtro\(^2\)** es la dimensión del filtro (por ejemplo, 3x3 o 5x5).
- **Número de canales de entrada** refiere a la profundidad del volumen de entrada; para la primera capa, serían los canales de color de la imagen (3 en este caso para RGB), y para las capas subsiguientes, sería el número de mapas de características generados por la capa anterior.
- **+1** representa el sesgo asociado con cada filtro.
- **Número de filtros** es la cantidad de mapas de características que la capa debe producir.

**La expresión de cálculo de los parámetros es: kernel * entradas + 1 * filtro**

**Cálculo capa 1:** (3x3x3+1)x100 = 2800

La capa 1 al haber aprendió 100 filtros los mismo se multiplican en la capa dos

**Cálculo capa 2:**  (3x3x100)+1x200 = 180.200

**Cálculo capa 3:**  (5x5x200)+1x300 = 1.500.300

**Cálculo capa 4:**  (5x5x300)+1x400 = 3.000.400

**Cálculo capa 5:**  (5x5x400)+1x500 = 5.000.500


### Total de Parámetros

El número total de parámetros en la red será la suma de los parámetros de todas las capas:

\[ 2800 + 180,200 + 1,500,300 + 3,000,400 + 5,000,500 = 9,683,200 \]

Por lo tanto, el número total de parámetros que tendríamos que estimar para esta CNN sería 9,683,200. Este número incluye tanto los pesos de los filtros en cada capa como los sesgos asociados a cada mapa de características.

2. Definición de la arquitectura red neuronal multicapa

```python
# Tenemos que importar Keras y funciones adicionales de sublibrerías de Keras
import keras # a su vez importa TensorFlow
from keras.models import Sequential # necesitamos la función de Sequential para inicializar los valores de los parámetros de las RNA
from keras.layers import Dense # y la función Dense que sirve para crear cada una de las capas intermedias (ocultas) de las RNA
```

```python
# Inicializamos la RNA con la función Sequential
rna = Sequential()
```

```python
# Vamos a tener 11 nodos en la capa de entrada ya que tenemos 11 variables independientes
x_train.shape
```

```python
# Añadimos las capas de entrada y una primera capa oculta utilizando la función Dense
# units = no. nodos que vamos a considerar en la primera capa oculta
# kernel_initializer = función de distribución para inicializar los pesos
# activation = función de activación elegida
# input_dim = no. nodos que vamos a considerar en la capa de entrada
rna.add(Dense(units = 6, kernel_initializer = "uniform",
                     activation = "relu", input_dim = 11))
```

```python
# Añadimos una segunda capa oculta
rna.add(Dense(units = 6, kernel_initializer = "uniform",  activation = "relu"))
```

```python
# Finalmente añadimos la capa de salida
rna.add(Dense(units = 1, kernel_initializer = "uniform",  activation = "sigmoid"))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Cargar Fashion-MNIST
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Normalización y Reshape
train_images = train_images / 255.0
test_images = test_images / 255.0
train_images = train_images.reshape((-1, 28, 28, 1))
test_images = test_images.reshape((-1, 28, 28, 1))

# Diseño de la CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),  # Corregido aquí
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compilar
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entrenar
history = model.fit(train_images, train_labels, epochs=10, validation_split=0.2)

# Validar
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'\nTest Accuracy: {test_acc}')


```

```python
import numpy as np
from tensorflow import keras
from sklearn.model_selection import KFold

def create_model():
    model = keras.Sequential([
        keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Flatten(),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Asegúrate de que train_images y train_labels están preprocesados (normalizados y reshaped)
num_folds = 5
kfold = KFold(n_splits=num_folds, shuffle=True)

fold_no = 1
acc_per_fold = []
loss_per_fold = []

# KFold proporciona índices de los datos para entrenar y probar en cada pliegue
for train, test in kfold.split(train_images, train_labels):
    model = create_model()
    print('------------------------------------------------------------------------')
    print(f'Training for fold {fold_no} ...')

    # Entrenamiento del modelo para el pliegue actual
    history = model.fit(train_images[train], train_labels[train],
                        batch_size=32, epochs=10, verbose=1)

    # Evaluación del modelo en el conjunto de prueba del pliegue actual
    scores = model.evaluate(train_images[test], train_labels[test], verbose=0)
    print(f'Score for fold {fold_no}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')
    acc_per_fold.append(scores[1] * 100)
    loss_per_fold.append(scores[0])

    fold_no += 1

# Imprimir los resultados promedio de todos los pliegues
print('------------------------------------------------------------------------')
print('Score per fold')
for i in range(len(acc_per_fold)):
    print(f'> Fold {i+1} - Loss: {loss_per_fold[i]} - Accuracy: {acc_per_fold[i]}%')
print('------------------------------------------------------------------------')
print('Average scores for all folds:')
print(f'> Accuracy: {np.mean(acc_per_fold)} (+- {np.std(acc_per_fold)})')
print(f'> Loss: {np.mean(loss_per_fold)}')
print('------------------------------------------------------------------------')
```

```python
import tensorflow as tf
from tensorflow import keras
from keras_tuner import RandomSearch, HyperParameters

def model_builder(hp):
    model = keras.Sequential()
    model.add(keras.layers.Conv2D(filters=hp.Choice('conv_filters', values=[32, 64]),
                                  kernel_size=(3, 3), activation='relu',
                                  input_shape=(28, 28, 1)))
    model.add(keras.layers.MaxPooling2D((2, 2)))
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(units=hp.Choice('dense_units', values=[64, 128]), activation='relu'))
    model.add(keras.layers.Dropout(rate=hp.Choice('dropout_rate', values=[0.2, 0.5])))
    model.add(keras.layers.Dense(10, activation='softmax'))

    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3])),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Inicializa el RandomSearch tuner
tuner = RandomSearch(
    model_builder,
    objective='val_accuracy',
    max_trials=16,  # Ajusta esto según tus necesidades
    executions_per_trial=1,
    directory='my_dir',
    project_name='kt_grid_search'
)

stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)

# Realiza la búsqueda
tuner.search(train_images, train_labels, epochs=10, validation_split=0.2, callbacks=[stop_early])

# Obtener los mejores hiperparámetros
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"Los mejores hiperparámetros son:\n Convolutional filters: {best_hps.get('conv_filters')}\n Dense units: {best_hps.get('dense_units')}\n Dropout rate: {best_hps.get('dropout_rate')}\n Learning rate: {best_hps.get('learning_rate')}")

# Construye el modelo con los mejores hiperparámetros
model = tuner.hypermodel.build(best_hps)

# Entrena el modelo
history = model.fit(train_images, train_labels, epochs=10, validation_split=0.2)
```

El objetivo es ajustar una red neuronal sobre el conjunto de datos CIFAR-10 para obtener una precisión superior al 89% en el conjunto de prueba. CIFAR-10 contiene 60.000 imágenes en color de 32x32 píxeles divididas en 10 clases: avión, automóvil, pájaro, gato, ciervo, perro, rana, caballo, barco y camión.

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

# Cargamos los datos
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalizamos los valores de los píxeles dividiéndolos por 255 para tener los valores entre 0-1
x_train, x_test = x_train / 255.0, x_test / 255.0

# aplicamos Codificación one-hot para las etiquetas
y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)

# Verificación de las dimensiones de los datos
print(f'x_train shape: {x_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'x_test shape: {x_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Definición del modelo
model = Sequential([
    # Primera capa convolucional:
    # 32 filtros, cada uno de tamaño 3x3, función de activación ReLU.
    # 'input_shape' define la forma de las entradas (imágenes): 32x32 píxeles, 3 canales (RGB).
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),

    # Primera capa de Max Pooling:
    # Reduce la dimensionalidad de la salida de la capa convolucional anterior, tomando el máximo de cada región 2x2.
    MaxPooling2D((2, 2)),

    # Segunda capa convolucional:
    # Incrementa el número de filtros a 64, permitiendo que la red aprenda características más complejas.
    Conv2D(64, (3, 3), activation='relu'),

    # Segunda capa de Max Pooling:
    # Continúa el proceso de reducción de dimensionalidad.
    MaxPooling2D((2, 2)),

    # Capa de aplanamiento:
    # Convierte los mapas de características 2D resultantes en un vector 1D para permitir el procesamiento por capas densas.
    Flatten(),

    # Primera capa densa (fully-connected):
    # 128 unidades/neuronas, función de activación ReLU.
    # Esta capa toma el vector aplanado y aprende las relaciones no lineales entre las características.
    Dense(128, activation='relu'),

    # Capa de Dropout:
    # Descarta aleatoriamente el 50% de las características durante el entrenamiento para prevenir el sobreajuste.
    Dropout(0.5),

    # Capa de salida:
    # 10 unidades/neuronas, una por cada clase posible (las 10 clases de CIFAR-10).
    # Función de activación 'softmax' para obtener probabilidades de cada clase.
    Dense(10, activation='softmax')
])

# Compilación del modelo:
# Optimizador Adam: Método de optimización basado en el gradiente estocástico que ajusta la tasa de aprendizaje.
# Pérdida: 'categorical_crossentropy', adecuada para clasificación multiclase.
# Métrica: 'accuracy', para evaluar el rendimiento del modelo durante el entrenamiento y las pruebas.
model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
# Importaciones necesarias
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Carga de datos
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)

# Función para crear el modelo
def create_model(dropout_rate=0.5, learning_rate=0.001):
    model = Sequential([
        Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
        BatchNormalization(),
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(dropout_rate),

        Conv2D(64, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(dropout_rate),

        Conv2D(128, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(dropout_rate),

        Flatten(),
        Dense(128, activation='relu'),
        Dropout(dropout_rate),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Entrenamiento del modelo con parámetros seleccionados
model = create_model(dropout_rate=0.4, learning_rate=0.0005)
checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy', mode='max')
early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=50,  # Número razonable de épocas
    batch_size=64,  # Tamaño de batch que equilibra tiempo de entrenamiento y uso de memoria
    callbacks=[checkpoint, early_stopping],
    verbose=1
)

# Evaluación del modelo
model.load_weights('best_model.h5')
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_accuracy*100:.2f}%")

# Visualización de la precisión y la pérdida
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```
